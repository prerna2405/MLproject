```python
def mann_dist(doc1_index, doc2_index):
	return sum([abs(user_dataset[doc1_index][i]-user_dataset[doc2_index][i]) for i in range(145)])

document_distance_matrix = []
for i in range(num_users):
	document_distance_matrix.append([mann_dist(i,j) for j in range(num_users)])

print(document_distance_matrix)
```

```python
## Take 50 as the threashold
## Finding the optimal value of k 

k = int(input("Enter the value of k: "))
num_docs = num_users
num_words = 145
num_docs_per_clusters = num_docs//k

docs = [i for i in range(num_docs)]
cluster = []
for i in range(k-1):
    cluster.append(random.sample(docs, num_docs_per_clusters))
    for j in range(num_docs_per_clusters):
        docs.remove(cluster[i][j])

cluster.append(docs)
print("Random clusters formed are ",cluster)

def find_centroid(doc_list):
    centroid = []
    for i in range(145):
        temp = 0
        for j in range(len(doc_list)):
            temp += user_dataset[doc_list[j]][i]
        centroid.append(round(temp/len(doc_list),2))
    return centroid


centroid_matrix = []
for i in range(len(cluster)):
        centroid_matrix.append(find_centroid(cluster[i]))


def centroid_doc_dist(doc_index, centroid_index):
    return sum([abs(user_dataset[doc_index][i]-centroid_matrix[centroid_index][i]) for i in range(num_words)])

for x in range(100):
    
    document_centroid_matrix = []

    for i in range(num_docs):
        document_centroid_matrix.append([round(centroid_doc_dist(i,j),2) for j in range(k)])
    #print("Centroid Matrix")
    #pprint(centroid_matrix)
    #print("Document Centroid Matrix")
    #pprint(document_centroid_matrix)

    del cluster[:]
    for _ in range(k):
        cluster.append([])
    
    #print(cluster)
    for i in range(num_docs):
        cluster[document_centroid_matrix[i].index(min(document_centroid_matrix[i]))].append(i)
    #print("Clusters\n",cluster)

    del centroid_matrix[:]
    for i in range(len(cluster)):
        centroid_matrix.append(find_centroid(cluster[i]))


for i in range(len(cluster)):
    for j in range(len(cluster[i])):
        cluster[i][j]+=1
    
print("Final Clusters are: ")
print(cluster)
```

```python
def psmin(k):
    new=[]
    for i in k:
        if (i!=0):
            new.append(i)
    mini= min(new)
    return mini


def get_probable_cuisine(i):
    cluster_cuisine = []
    prob = [0 for k in range(145)]
    for j in range(len(cluster[i])):
        for x in range(145):
            prob[x]=prob[x]+user_dataset[cluster[i][j]-1][x]
    
    total_sum = sum(prob)
    for _ in range(145):
        prob[_] = prob[_]/total_sum
    
    threashold = (max(prob) + psmin(prob))/2
    
    for _ in range(145):
        if(prob[_] >= threashold):
            cluster_cuisine.append(list(cuisines)[_])
    
    return cluster_cuisine
    
        
    

my_dict = {}
for i in range(k):
    my_dict[i] = list(set(get_probable_cuisine(i)))
    
print(my_dict)
```
